1.  Explique qué es una heurística en el problema del laberinto,
    > proponga dos heurísticas y explique cómo pueden ayudar a resolver
    > mejor el problema del laberinto.

Brevemente, una heurística es una función que clasifica las alternativas
en los algoritmos de búsqueda en cada paso de bifurcación en función de
la información disponible para decidir qué rama seguir. Por ejemplo,
puede aproximar la solución exacta. Ahora en el caso del laberinto, una
heurística es una función que estima qué tan cerca está un estado a la
meta. La primera heurística podría ser hallar la hipotenusa del estado
hasta la meta. La segunda opción puede ser fragmentar los caminos hasta
llegar a la meta siendo el número de bloques restantes a la meta. Estos
pueden resolver mejor el problema del laberinto pues siempre busca la
distancia más cercana al punto de llegada.

1.  Explique brevemente el algoritmo de búsqueda greedy. En particular,
    > su estrategia de exploración y la representación de la frontera.

El algoritmo de búsqueda greedy entra en la categoría de algoritmos de
"búsqueda mejor primero", es decir, es un paradigma algorítmico que
construye una solución pieza por pieza, eligiendo siempre la siguiente
pieza que ofrece el beneficio más obvio e inmediato. Luego, el greedy no
usa el conocimiento previo para expandirse de nodo a nodo. En el greedy
todos los nodos en el borde (o franja o frontera) se mantienen en la
memoria, y los nodos que ya se han expandido no necesitan almacenarse en
la memoria y, por lo tanto, se pueden descartar.



1.  Explique brevemente el algoritmo A\*. En particular, la estrategia
    > de búsqueda y la representación de la frontera.

El algoritmo A\* también entra en la categoría de algoritmos de
"búsqueda mejor primero", usando una función f que estima el costo total
el camino desde el n hasta la meta, representado
por$\ \ f(n)\  = \ \ g(n)\  + h(n)$ donde h es una función heurística, g
el costo máximo de n hasta la meta, A\* utiliza una función heurística
admisible, lo que esencialmente significa que A\* es óptimo, es decir,
siempre encuentra el camino óptimo entre el nodo de inicio y el nodo de
destino. Sin embargo, A\* necesita mantener todos los nodos en la
memoria mientras busca, no solo los que se encuentran en la frontera,
porque A\*, esencialmente, realiza una"búsqueda exhaustiva". Ahora en el
grafo de búsqueda: A\* es óptimo si la heurística es consistente. Por lo
tanto, la consistencia implica admisibilidad.En general, la mayoría de
las heurísticas admisibles naturales tienden a ser consistentes.



1.  Un algoritmo de búsqueda se dice admisible si tiene garantía de
    > retornar una solución óptima. Si la función heurística utilizada
    > por A\* es admisible, entonces A\* es admisible. Explique que es
    > una heurística admisible, y pruebe la afirmación previa.

Una heurística h es admisible si: $0\  < \ h(n)\  \leq \ h*(n)$

Donde h\*(n) es el costo real a la meta más cercana, además,
h(meta)=0.En el caso del laberinto al punto de llegada.

Entonces para probar que si la función heurística utilizada por A\* es
admisible, entonces A\* es admisible supongamos que aparece un nodo
objetivo óptimo G2 en la periferia, sea C∗ el costo de la solución
óptima, entonces como G2 es óptimo y (G2)=0, sabemos
$\ \ f(G2)\  = \ \ g(h2)\  + h(G2) = g(G2)\  > \ C*$

Si h(n) no sobreestima el costo de completar el camino de la solución,
entonces sabemos que,

$\ \ f(n)\  = \ \ g(n)\  + h(n) \leq \ C*$ Ahora hemos demostrado que
$\ \ f(n’)\  \leq C* \leq f(G2)\ \ $

por lo que G2 no se expandirá y A\* debe devolver una solución óptima.
Llegando a que A\* es admisible.

1.  Explique en qué consiste una heurística consistente, muestre un
    > ejemplo de una heurística admisible pero no consistente, y pruebe
    > el resultado de Dechter y Pearl.

1\. Una heurística es consistente si para cada nodo n, cada sucesor n'
de n generado por cualquier acción a,
$\ \ h(n)\  \leq c(n,a,n’)\  + \ h(n’)$

Ahora si h es consistente, se tendría que

$\ \ f(n’)\  = \ \ g(n’)\  + h(n’)$

$ = \ \ g(n)\ \  + c(n,a,n’)\  + \ h(n’)$

$\text{ } \geq \ \ g(n)\ \  + h(n)$

$= \ \ f(n)$

2\. Ejemplo de heurística admisible pero no
consistente

-   Admisibilidad:

h(A) \<= 4

h(C) \<= 3

h(G) \<= 0

Vemos que las heurísticas que tenemos son admisibles porque se

tiene que h(n) \<=h\*(n) para cada nodo n donde h\* es el costo real

-   Consistencia

Para que sea consistente se tiene que h(n) \<= cost(n, c) + h(c) donde
el nodo c es hijo del nodo n. Teniéndose lo siguiente:

h(A) \<= 1 + h(C)

h(C) \<= 3 + h(G) = 3

Pero se puede llegar a una inconsistencia si:

h(A) \> 1 + h(C)

h(C) \<= 3 h(G) = 0

Manteniéndose la admisibilidad pero fallando en la consistencia de la
heurística.
